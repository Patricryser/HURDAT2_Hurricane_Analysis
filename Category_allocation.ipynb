## Atlantic Storm Category allocation

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("atlantic_storms_trans_bearing.csv",parse_dates=['date'], index_col=[0])
```

```python
df.head()
```

### Adding the Hurricane Category

```python
#Add a new column in our dataframe df named Category
df['category'] = np.nan
for i in df.index:
    if ((df.maximum_sustained_wind_knots.iloc[i]>=64) and (df.maximum_sustained_wind_knots.iloc[i]<= 82) ):
        df.at[i,'category'] = int(1)
    if ((df.maximum_sustained_wind_knots.iloc[i]>=83) and (df.maximum_sustained_wind_knots.iloc[i]<=95) ):
        df.at[i,'category'] = int(2)
    if ((df.maximum_sustained_wind_knots.iloc[i]>=96) and (df.maximum_sustained_wind_knots.iloc[i]<=112) ):   
        df.at[i,'category'] = int(3)
    if ((df.maximum_sustained_wind_knots.iloc[i]>=113) and (df.maximum_sustained_wind_knots.iloc[i]<=136) ):
        df.at[i,'category'] = int(4)
    if (df.maximum_sustained_wind_knots.iloc[i]>=137):
        df.at[i,'category'] = int(5)  
    else:
        df.at[i,'category'] = int(0)
```
